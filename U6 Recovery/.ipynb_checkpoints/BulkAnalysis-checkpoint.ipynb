```r
bulk = read.table('BCR/U6Bulk.tab', header = TRUE, sep = '\t', stringsAsFactors = FALSE) 
```

```r
library(ggplot2)
library(dplyr)
UMAP_theme <- theme_bw()+theme(axis.text.y = element_blank(), 
                               axis.text.x = element_blank(), 
                               axis.ticks.x= element_blank(),
                               axis.ticks.y= element_blank(),
                               axis.title.x= element_blank(),
                               axis.title.y= element_blank(),
                               strip.text.x = element_text(size = 8),
                               panel.grid.major = element_blank(),
                               panel.grid.minor = element_blank(),
                               panel.border = element_blank(),
                               panel.background = element_rect(colour = "black", fill = NA,size=.355),
                               legend.position = "none",
                               plot.title = element_text(hjust = 0.5, size = 8))

remove_grid = theme_bw() + theme( panel.grid.major = element_blank(),
                               panel.grid.minor = element_blank())

Axis_themes <- theme(plot.title = element_text(size = 8),
                     axis.title = element_text(size = 8), 
                     axis.text = element_text(size = 6),
                     axis.text.x = element_text(size = 6),
                     legend.text = element_text(size =6),
                     legend.title = element_text(size = 8),
                     strip.text.x = element_text(size = 8), 
                    axis.line = element_line(size = .355), 
                    axis.ticks = element_line(size = .177))
```

```r
cell_consensus_hc = readRDS('hc_consensus.ta')

hc_bulk = bulk %>% filter(LOCUS == 'IGH', FUNCTIONAL, Sample %in% c('D23-1208', 'D23-1209')) %>% filter(CONSCOUNT1 > 5, CONSCOUNT2 > 5)
hc_sc = hc %>% filter(FUNCTIONAL)

all_hc = merge(hc_bulk, hc_sc, all = TRUE)
```

```r
library(shazam)
```

```r
lc_bulk = bulk %>% filter(LOCUS != 'IGH', FUNCTIONAL,! Sample %in% c('D23-1208', 'D23-1209'))
library(shazam)
library(alakazam)
hc = readRDS('hc.RDS')
lc = readRDS('lc.RDS')
```

```r
options(repr.plot.width = 3,repr.plot.height = 3, repr.plot.res = 300)
bulk_genes = countGenes(hc_bulk , gene = 'V_CALL')%>% filter(seq_count > 0
                                                                                                   )#%>% filter(CDR3_IMGT != 'GCACACCTATACTATGGTTCGGGGAGTTATCAATTTGACTAC'), gene = 'V_CALL') 
sc_genes = countGenes(hc ,gene = 'V_CALL') %>% filter(seq_count > 0)
sc_genes$bulk_freq = bulk_genes$seq_freq[match(sc_genes$gene, bulk_genes$gene)]
sc_genes$bulk_freq = bulk_genes$seq_freq[match(sc_genes$gene, bulk_genes$gene)] 
ggplot(sc_genes, aes(x = seq_freq, y = bulk_freq)) + geom_point(size = .3) + scale_x_log10(limits = c(1/10000, 1/10)) + scale_y_log10(limits = c(1/10000, 1/10))  + ggpubr::stat_cor(method = 'spearman',size = 2.167) + theme_classic() + 
labs(x = 'Frequency in single-cell data', y = 'Frequency in bulk RNA-seq') + Axis_themes + geom_abline(color = 'red', linetype = 'dashed', size = .155) + ggtitle('Heavy chain') + theme(plot.title = element_text(hjust = .5)) + 
geom_text(aes(label = gene), size = 2.11676)

```

```r
options(repr.plot.width = 3,repr.plot.height = 3, repr.plot.res = 300)
bulk_genes = countGenes(hc_bulk %>% filter(!duplicated(SEQUENCE_IMGT)) , gene = 'V_CALL')%>% filter(seq_count > 0
                                                                                                   )#%>% filter(CDR3_IMGT != 'GCACACCTATACTATGGTTCGGGGAGTTATCAATTTGACTAC'), gene = 'V_CALL') 
sc_genes = countGenes(hc %>% filter(!duplicated(SEQUENCE_IMGT)),gene = 'V_CALL') %>% filter(seq_count > 0)
sc_genes$bulk_freq = bulk_genes$seq_freq[match(sc_genes$gene, bulk_genes$gene)]
sc_genes$bulk_freq = bulk_genes$seq_freq[match(sc_genes$gene, bulk_genes$gene)] 
g1 = ggplot(sc_genes, aes(x = seq_freq, y = bulk_freq)) + geom_point(size = .3) + scale_x_log10(limits = c(1/10000, 1/10)) + scale_y_log10(limits = c(1/10000, 1/10))  + ggpubr::stat_cor(method = 'spearman',size = 2.167) + theme_classic() + 
labs(x = 'Frequency in single-cell data', y = 'Frequency in bulk RNA-seq') + Axis_themes + geom_abline(color = 'red', linetype = 'dashed', size = .155) + ggtitle('Heavy chain') + theme(plot.title = element_text(hjust = .5))

```

```r
options(repr.plot.width = 3,repr.plot.height = 3, repr.plot.res = 300)
bulk_genes = countGenes(lc_bulk %>% filter(!duplicated(SEQUENCE_IMGT)) , gene = 'V_CALL')%>% filter(seq_count > 0
                                                                                                   )#%>% filter(CDR3_IMGT != 'GCACACCTATACTATGGTTCGGGGAGTTATCAATTTGACTAC'), gene = 'V_CALL') 
sc_genes = countGenes(lc %>% filter(!duplicated(SEQUENCE_IMGT)),gene = 'V_CALL') %>% filter(seq_count > 0)
sc_genes$bulk_freq = bulk_genes$seq_freq[match(sc_genes$gene, bulk_genes$gene)]
sc_genes$bulk_freq = bulk_genes$seq_freq[match(sc_genes$gene, bulk_genes$gene)]

g2 = ggplot(sc_genes, aes(x = seq_freq, y = bulk_freq)) + geom_point(size = .3) + scale_x_log10(limits = c(1/10000, 1/10)) + scale_y_log10(limits = c(1/10000, 1/10))  + ggpubr::stat_cor(method = 'spearman', size = 2.1167) + theme_classic() +  

labs(x = 'Frequency in single-cell data', y = 'Frequency in bulk RNA-seq') + Axis_themes + geom_abline(color = 'red', linetype = 'dashed', size = .155) + 
 ggtitle('Light chain') + theme(plot.title = element_text(hjust = .5))
```
